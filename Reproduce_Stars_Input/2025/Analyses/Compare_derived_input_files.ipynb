```python
jul_dates = pd.read_csv(stars_dir + "2024/measure_dates/2023_07_measure_dates.csv")
oct_dates = pd.read_csv(stars_dir + "2024/measure_dates/2023_10_measure_dates.csv")
nov_dates = pd.read_csv(stars_dir + "2024/measure_dates/2023_11_measure_dates.csv")
jan_dates = pd.read_csv(stars_dir + "2024/measure_dates/2024_01_measure_dates.csv")

actual_dates = jan_dates.copy(deep=True)
jan_dates.head(jan_dates.shape[0])
```

```python
def smape(A, F):
    with np.errstate(divide='ignore', invalid='ignore'):
        tmp = 2 * np.abs(F-A) / (np.abs(A) + np.abs(F))
    tmp[np.isnan(tmp)] = 0
    return np.sum(tmp) / len(tmp) * 100


def compare_dfs(exp_df, obs_df, mo_yr, exp_dates, obs_dates):
    
    print(mo_yr)
    
    for lab in list(exp_df):
        if lab == 'PROVIDER_ID':
            continue
        
        else:
            # Perform Comparisons
            
            # 0. Dates
            
            
            
            # 1. print no. of NaNs (obs vs exp)
            nans_exp = exp_df[lab].isna().sum()
            nans_obs = obs_df[lab].isna().sum()
            print(f'Number of NaNs in observed ({lab}): {nans_obs}')
            print(f'Number of NaNs in expected ({lab}): {nans_exp}')
            
            # 2. Print SMAPE
            # Drop NaNs for the difference calculation
            valid_exp = exp_df[lab].dropna()
            valid_obs = obs_df[lab].dropna()
            common_index = valid_exp.index.intersection(valid_obs.index)
            valid_exp = valid_exp.loc[common_index]
            valid_obs = valid_obs.loc[common_index]
            
            if valid_exp.empty or valid_obs.empty:
                print(f'No valid data to compare for {lab}')
                print('\n')
                continue
            
            # Calculate SMAPE
            smape_value = smape(valid_exp, valid_obs)
            
            print(f'Symmetric Mean Absolute Percentage Error (SMAPE) for {lab}: {smape_value:.2f}%')
            
            
            print('\n')
```

```python
compare_dfs(df_2024, jul_df, 'July 2023', jul_dates, actual_dates)
```

```python
compare_dfs(df_2024, nov_df, 'Nov 2023', nov_dates, actual_dates)
```

```python
compare_dfs(df_2024, jan_df, 'Jan 2024', jan_dates, actual_dates)
```
